```python
import pandas as pd
from pathlib import Path
import random
```

```python
DATA_DIR = Path('dataset/discovery')
```

```python
# Load interactions dataset

# Load approximately p percent of dataset
p = 0.01

interactions_df = pd.read_csv(
    DATA_DIR / 'interactions.csv',
    skiprows=lambda i: i>0 and random.random() > p,
)
interactions_df
```

```python
# Take all the items

unique_items = interactions_df['ITEM_ID'].unique()
print(f'size of items: {unique_items.size}')
```

```python
# Load items dataset

items_df = pd.read_csv(DATA_DIR / 'items.csv')
items_df
```

```python
# Filter items dataset based on items from interactions dataset

items_df = items_df.loc[items_df['ITEM_ID'].isin(unique_items), :].reset_index(drop=True)
items_df
```

```python
# Rename the column names

interactions_df = interactions_df.rename(columns={
    'USER_ID': 'user_id:token',
    'ITEM_ID': 'item_id:token',
    'EVENT_TYPE': 'event_type:token',
    'EVENT_VALUE': 'event_value:float',
    'LOCATION': 'location:token',
    'ASSET_LIST_INDEX': 'asset_list_index:token',
    'ASSET_CAROUSEL_INDEX': 'asset_carousel_index:token',
    'TIMESTAMP': 'timestamp:float'
})

interactions_df.head()
```

```python
items_df = items_df.rename(columns={
    'ITEM_ID': 'item_id:token',
    'TYPE': 'type:token',
    'HASHTAGS': 'hashtags:token_seq',
    'SOURCE': 'source:token',
    'TITLE': 'title:token_seq',
    'DURATION': 'duration:float',
    'CREATION_TIMESTAMP': 'creation_timestamp:float'
})

items_df.head()
```

```python
# Save the files in Atomic format
interactions_df.to_csv(DATA_DIR / 'discovery.inter', sep='\t', index=False)
items_df.to_csv(DATA_DIR / 'discovery.item', sep='\t', index=False)
```
